```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # to visualize data
import seaborn as sns
```

```python
df= pd.read_csv("WeatherHistory.csv") # extracting data using pandas
```

```python
df.head()
```

```python
df.info() # this gives the summary of the dataset
```

The dataset has 96453 entries with total 12 columns. In the 12 columns, 4 are objects and 8 are numerical data types.

```python
df.describe()
```

```python
pd.isnull(df).sum() # check for null values
```

we can see that there are 517 null values in Precip Type column

```python
df.dropna(inplace=True)  # drop null values
```

```python
pd.isnull(df).sum()
```

Now there are no null values in the dataset.

Exploratory Data Analysis

```python
sns.set(rc={'figure.figsize':(16,5)})
plt.xticks(rotation=90)
sns.barplot(x='Summary',y='Temperature (C)',data=df)
```

Highest temperature is of Dry weather whereas lowest temperature is of fooggy.

```python
sns.set(rc={'figure.figsize':(16,5)})
plt.xticks(rotation=90)
sns.lineplot( x='Summary', y='Humidity', hue='Precip Type',data=df)
```

```python
plt.xticks(rotation=90)
ax= sns.countplot(x='Summary', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

Most of the days are partly cloudy followed by mostly cloudy, overcast and clear.

```python
sns.set(rc={'figure.figsize':(12,5)})
sns.barplot( x='Precip Type', y='Visibility (km)',data=df)
```

Visibility is low during snow compared to the rainy day.

```python
sns.set(rc={'figure.figsize':(12,5)})
sns.scatterplot( x='Temperature (C)', y='Wind Speed (km/h)',hue='Precip Type',data=df)
```

* Wind speed is High on rainy days and low during snow.
* Temperature is High during rains and low during snow.

```python
plt.xticks(rotation=90)
bx= sns.countplot(x='Precip Type', data=df)
for bars in bx.containers:
    bx.bar_label(bars)
```

There are two Precip types Rain and snow. Most of the days it Rains.

```python
sns.barplot( x='Precip Type', y='Pressure (millibars)',hue='Summary',data=df)
```

Pressure is mostly high when there is rain.

```python
sns.barplot( x='Precip Type', y='Humidity',data=df)
```

Humidity is High when there is Snow.
